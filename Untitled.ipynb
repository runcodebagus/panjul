```python
df = pd.read_excel('data.xlsx')
df
```

```python
df.drop(labels=['No.', 'Nama Instagram', 'Tanggal Posting', 'Nama Akun IG  Artis/Selebgram', 'Unnamed: 6', 'Unnamed: 7', 'Kategori'], axis=1, inplace =True)
df
```

```python
# Membersihkan teks dari karakter khusus dan mengonversi ke huruf kecil
def preprocess_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    return text

df['Komentar'] = df['Komentar'].apply(preprocess_text)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words=stop_words) # stopwords adalah list stopword Bahasa Indonesia
```
